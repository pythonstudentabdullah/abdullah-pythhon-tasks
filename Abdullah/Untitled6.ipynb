```python
import tkinter as tk
from tkinter import*

root=Tk()
root.title("Registration")
root.geometry("600x470")
root.resizable(False,False)

def register():
    print("registered")



Label(root,text="Registration Form",font="arial 25").pack(pady=50)

Label(text="Name",font=23).place(x=100,y=150)
Label(text="Email",font=23).place(x=100,y=200)
Label(text="Gender",font=23).place(x=100,y=250)
Label(text="Programming",font=23).place(x=100,y=300)

#entry
nameValue=StringVar()
emailValue=StringVar()
genderValue=StringVar()
programmingValue=StringVar()
nameEntry=Entry(root,textvariable=nameValue,width=30,bd=2,font=2)
emailEntry=Entry(root,textvariable=emailValue,width=30,bd=2,font=2)
genderEntry=Entry(root,textvariable=genderValue,width=30,bd=2,font=2)
programmingEntry=Entry(root,textvariable=programmingValue,width=30,bd=2,font=2)


nameEntry.place(x=200,y=150)
emailEntry.place(x=200,y=200)
genderEntry.place(x=200,y=250)
programmingEntry.place(x=200,y=300)

#checkbutton
checkValue=IntVar
checkbtn=Checkbutton(text="remember me?",variable=checkValue)
checkbtn.place(x=200,y=340)

Button(text="Register",font=20,width=11,height=2,command=register).place(x=250,y=380)



root.mainloop()
```

```python
import matplotlib.pyplot as plt

days=["Sunday","Tuesday","Wednesday","Thursday","Friday","Saturday","Monday",]

numbers=[25,13,5,10,6,30,6]

plt.plot(days,numbers,color="yellow",linewidth=5,marker="*",markerfacecolor="blue",markersize=20
        ,linestyle="dashed")

plt.xlabel("Days")

plt.ylabel("Number Of Push-ups")

plt.title("My Push-ups Graph")

plt.show()
```
